## Étape 7 — Export de la table finale

Export CSV de la table d'expérience avec : `age`, `E_x`, `D_x`,
`q_x_brut`, `q_x_lisse`, `q_x_ref`, `SMR_local`.

```python
import os
output_cols = ['age','E_x','D_x','q_x_brut','q_x_lisse','q_x_ref','SMR_local']
table_finale = table[output_cols].copy()
for c in ['q_x_brut','q_x_lisse','q_x_ref','SMR_local']:
    table_finale[c] = table_finale[c].round(6)
base = os.path.splitext(os.path.basename(FILE_PATH))[0] if FILE_PATH else 'synthetique'
OUTPUT_PATH = 'table_mortalite_{}_{}.csv'.format(base, SEXE)
table_finale.to_csv(OUTPUT_PATH, index=False)
n_cov = table_finale['E_x'].gt(0).sum()
qx_min = table_finale['q_x_lisse'].dropna().min()
qx_max = table_finale['q_x_lisse'].dropna().max()
age_min = table_finale.loc[table_finale['q_x_lisse']==qx_min,'age'].iloc[0] if n_cov>0 else '?'
age_max = table_finale.loc[table_finale['q_x_lisse']==qx_max,'age'].iloc[0] if n_cov>0 else '?'
print('Table exportee     : ' + OUTPUT_PATH)
print('Ages couverts      : {:}-{:} ({:} ages)'.format(
      int(table_finale['age'].min()), int(table_finale['age'].max()), n_cov))
print('qx lisse min       : {:.6f} (age {:})'.format(qx_min, age_min))
print('qx lisse max       : {:.6f} (age {:})'.format(qx_max, age_max))
print('SMR global         : {:.4f}'.format(SMR))
print()
print('Extrait de la table finale :')
print(table_finale.dropna(subset=['q_x_lisse']).tail(10).to_string(index=False))
```
