## Étape 5 — Visualisation et contrôle graphique

Compare les taux bruts, lissés et la table de référence TH/TF 00-02.

```python
ages_plot = table['age'].values
q_ref_plot = np.array([qx_ref(a) for a in ages_plot])
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 5))
t_valid = table.dropna(subset=['q_x_brut'])
ax1.scatter(t_valid['age'], t_valid['q_x_brut'], s=25, alpha=0.6,
            label='Taux bruts', color='steelblue', zorder=2)
t_lisse = table.dropna(subset=['q_x_lisse'])
ax1.plot(t_lisse['age'], t_lisse['q_x_lisse'], color='crimson',
         linewidth=2, label='Lissage WH (lambda={:})'.format(LAMBDA_WH), zorder=3)
ax1.set_yscale('log'); ax1.set_xlabel('Age'); ax1.set_ylabel('qx (log)')
ax1.set_title('Taux bruts vs lisses'); ax1.legend(); ax1.grid(True, alpha=0.3)
ax2.plot(t_lisse['age'], t_lisse['q_x_lisse'], color='crimson',
         linewidth=2, label='Portefeuille lisse')
ax2.plot(ages_plot, q_ref_plot, color='navy', linewidth=2,
         linestyle='--', label='Reference TH/TF 00-02 ({:})'.format(SEXE))
ax2.set_yscale('log'); ax2.set_xlabel('Age'); ax2.set_ylabel('qx (log)')
ax2.set_title('Comparaison avec reference'); ax2.legend(); ax2.grid(True, alpha=0.3)
plt.tight_layout(); plt.show()
print('Graphiques generes.')
print('Age min avec donnees : {:}'.format(int(t_valid['age'].min())))
print('Age max avec donnees : {:}'.format(int(t_valid['age'].max())))
print('Ages avec lissage    : {:}'.format(len(t_lisse)))
```
